```
from konlpy.tag import Okt


# voc = '형태소 란 문장을 구성하는 의미 요소 중 가장 작은 단위를 말한다.'
okt_pos = Okt().pos(a, norm=True, stem=True)   # 형태소 분석 ( norm : 정규화 )


okt_filtering = [x for x, y in okt_pos if y in ['Noun', 'Adjective', 'Verb']]
print(okt_filtering)              # 불용어 처리 - 명사, 형용사, 동사만 선별


```

```
okt_filtering
```

```
stop_words = '하다 이다 있다 인천 잇다 가다 때 보다 기다 위 수 곳'
stop_words = stop_words.split(' ')
print(stop_words)

```

```

result = []
for w in okt_filtering: 
    if w not in stop_words: 
        result.append(w) 

print(result)
```

```
from collections import Counter
incheon_counter = Counter(result)
top_incheon = dict(incheon_counter.most_common(100))
top_incheon
```

```
import matplotlib.pyplot as plt 
import pandas as pd
import numpy as np

#단어 빈도 시각화
plt.rcParams['font.size'] = 12

y_pos = np.arange(len(top_incheon))

plt.figure(figsize=(12,24))
plt.barh(y_pos,top_incheon.values())
plt.title('Word Count')
plt.yticks(y_pos,top_incheon.keys())
plt.show()
```

```
import squarify
plt.rcParams['figure.figsize'] = (14,14)
plt.rcParams['font.size'] = 20

norm = mpl.colors.Normalize(vmin = min(top_incheon.values()), 
                            vmax = max(top_incheon.values()))
colors = [mpl.cm.Reds(norm(value)) for value in top_incheon.values()]

squarify.plot(label = top_incheon.keys(),
              sizes= top_incheon.values(),
              color = colors,
              alpha =.7);
```

```
from wordcloud import WordCloud
wc = WordCloud(background_color = 'white', font_path = './font/NanumBarunGothic.ttf')
wc.generate_from_frequencies(top_incheon)

figure = plt.figure(figsize=(14,14))
ax = figure.add_subplot(1,1,1)
ax.axis('off')
ax.imshow(wc)
plt.show()
```
