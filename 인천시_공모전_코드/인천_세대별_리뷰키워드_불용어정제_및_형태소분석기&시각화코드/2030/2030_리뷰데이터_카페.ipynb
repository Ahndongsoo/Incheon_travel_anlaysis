```
stop_words = '넓다 해변 공장 국적 한번 보이다 자리 물건 소품 값 앞 추천 존 말다 그냥 매우 주말 테이블 찍기 쏘다 받다 옛날 안잗 왕리 내부 별로 하나 테라스 망고 식사 가족 함 요 그렇다 갯벌 다시 방직 카페 기다리다 먹다 맛 떡볶이 물회 대기 강된장 포장 회 집 시키다 쌈 밥 김밥 줄 메뉴 주문 비 들어가다 진짜 떡 번호 국물 조금 느낌 싶다 듯 들다 번 가능하다 거리 이용 바람 처음 타다 코스 루지 월미도 바다 비싸다 개 올라가다 내려오다 탈 아쉽다 좀 장화 주기 작다 나오다 근처 먹이 코로나 주다 화장실 꼭 승마 가보다 애 분 보내다 체험 아이 많다 나비 고인돌 썰매 자연사 우주 입장료 강화도 시설 방문 아니다 공룡 가격 야외 토끼 강화 인천대 좋아하다 봉 정도 크다 더 좋다 무료 송도 전망대 층 공연 주차 볼 파크 센트럴 한눈 주차장 야경 뷰 장소 건물 경치 날 전망 눈 운영 호수 되어다 등 않다 사람 주변 생각 것 시간 같다 타고 정말 없다 오다 조개 숲 양 떼 자다 공원 목장 되다 산책 하다 이다 있다 인천 잇다 가다 때 보다 기다 위 수 곳'
stop_words = stop_words.split(' ')
print(stop_words)
```

```

result = []
for w in okt_filtering: 
    if w not in stop_words: 
        result.append(w) 

print(result)
```

```
from collections import Counter
incheon_counter = Counter(result)
top_incheon = dict(incheon_counter.most_common(100))
top_incheon
```
