```
from konlpy.tag import Okt


# voc = '형태소 란 문장을 구성하는 의미 요소 중 가장 작은 단위를 말한다.'
okt_pos = Okt().pos(a, norm=True, stem=True)   # 형태소 분석 ( norm : 정규화 )


okt_filtering = [x for x, y in okt_pos if y in ['Noun', 'Adjective', 'Verb']]
print(okt_filtering)              # 불용어 처리 - 명사, 형용사, 동사만 선별


```

```
stop_words = '싶다 듯 들다 번 가능하다 거리 이용 바람 처음 타다 코스 루지 월미도 바다 비싸다 개 올라가다 내려오다 탈 아쉽다 좀 장화 주기 작다 나오다 근처 먹이 코로나 주다 화장실 꼭 승마 가보다 애 분 보내다 체험 아이 많다 나비 고인돌 썰매 자연사 우주 입장료 강화도 시설 방문 아니다 공룡 가격 야외 토끼 강화 인천대 좋아하다 봉 정도 크다 더 좋다 무료 송도 전망대 층 공연 주차 볼 파크 센트럴 한눈 주차장 야경 뷰 장소 건물 경치 날 전망 눈 운영 호수 되어다 등 않다 사람 주변 생각 것 시간 같다 타고 정말 없다 오다 조개 숲 양 떼 자다 공원 목장 되다 산책 하다 이다 있다 인천 잇다 가다 때 보다 기다 위 수 곳'
stop_words = stop_words.split(' ')
print(stop_words)
```

```

result = []
for w in okt_filtering: 
    if w not in stop_words: 
        result.append(w) 

print(result)
```

```
from collections import Counter
incheon_counter = Counter(result)
top_incheon = dict(incheon_counter.most_common(100))
top_incheon
```
